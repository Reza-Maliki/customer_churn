```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
df=pd.read_csv("churn.csv")
df.isna().any()
#df.region_category.value_counts()
mode_region_category=df.region_category.mode()[0]
df.region_category.fillna(mode_region_category,inplace = True)
# df.isna().any()
```

```python
# df.preferred_offer_types.value_counts()
# df.preferred_offer_types.isna().sum()
df.preferred_offer_types.fillna("any one",inplace=True)
# df.isna().any()
```

```python
# df.points_in_wallet.value_counts()
# print(df.points_in_wallet.max())
# print(df.points_in_wallet.min())
# print(df.points_in_wallet.mean())
df.points_in_wallet.fillna(df.points_in_wallet.mean(),inplace=True)
# df.isna().any()
```

```python
#change numeric & date values to categorical data

df["cat_ages"]=pd.cut(df.age,bins=[0,18,30,60,65],labels=["teen","young","Middle-aged","old"])

df['joining_date'] = pd.to_datetime(df['joining_date'], format="%m/%d/%Y")
df["cat_joining_date"] = np.select([df['joining_date'] <= "2015/12/15", ("2016/12/15">=df['joining_date'])&(df['joining_date']>= "2015/12/16"),("2017/12/15">=df['joining_date'])&(df['joining_date']>= "2016/12/16")], ['2015','2016','2017'], '2020')
df["cat_joining_date"]=pd.Categorical(df.cat_joining_date)
# print(df["joining_date"].dt.year.value_counts())

# df.head(10)
df['last_visit_time_hour'] = pd.to_datetime(df['last_visit_time'], format="%H:%M:%S")
df['last_visit_time_hour'].dt.hour.head()
# df.info()
# df.memory_usage()
#df.loc[df.membership_category.isin(["Basic"])]
df
```

```python
#df.drop(["age","joining_date","referral_id"],axis=1,inplace=True)
df["cat_ages"]=pd.Categorical(df.cat_ages)
```

```python
#machine learning started:
feature=df[['cat_ages', 'gender', 'region_category', 'membership_category', 'feedback', 'cat_joining_date','points_in_wallet',"preferred_offer_types"]]

target=df['churn_risk_score']

feature=pd.get_dummies(feature,columns=['cat_ages', 'gender', 'region_category', 'membership_category', 'feedback', 'cat_joining_date','preferred_offer_types'])
feature.transpose()
feature_train, feature_test, target_train, target_test = train_test_split(feature, target, random_state=101)

rf_model = RandomForestClassifier(n_estimators=50, max_features="auto", random_state=44)
rf_model.fit(feature_train,  target_train.values.ravel())

predictions = rf_model.predict(feature_test)
print(target_test)


print("accuracy:", accuracy_score(target_test, predictions))
print("precision:", precision_score(target_test, predictions))
print("recall:", recall_score(target_test, predictions))

```
